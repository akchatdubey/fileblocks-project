```python
import matplotlib.pyplot as plt
subjects = ['Math', 'Physics', 'Chemistry', 'Biology', 'English']
marks = [85, 90, 78, 88, 92]
plt.bar(subjects, marks, color='lightgreen')
plt.xlabel('Subjects')
plt.ylabel('Marks Obtained')
plt.title('Student Marks in Different Subjects')
plt.legend("subjects",loc="upper right")
plt.show()
```

```python
import matplotlib.pyplot as plt
labels = ['Math', 'Physics', 'Chemistry', 'Biology', 'English']
marks = [85, 90, 78, 88, 92]
colors = ['gold', 'lightcoral', 'lightskyblue', 'lightgreen', 'violet']
plt.pie(marks, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.title('Student Marks Distribution')
plt.axis('equal')
plt.legend(title="MARKS:",loc="upper right")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.linspace(-10, 10, 400)  
y = x**2
plt.plot(x, y, label="y = x^2", color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of y = x^2')
plt.grid(True)
plt.legend()
plt.show()
```
