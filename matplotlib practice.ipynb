```python
import matplotlib.pyplot as plt
marks = [85, 92, 76, 81, 95, 67, 88, 74, 90, 82, 77, 85, 93, 78, 89, 91, 76, 84, 70, 69]
plt.hist(marks, bins=10, color='skyblue', edgecolor='black')
plt.xlabel('Marks')
plt.ylabel('Number of Students')
plt.title('Marks Distribution Histogram')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)  
y1 = np.sin(x)  
y2 = np.cos(x)  
y3 = np.tan(x)  
y4 = np.exp(x / 5)  
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes[0, 0].plot(x, y1, 'b')
axes[0, 0].set_title('Sine Function')
axes[0, 0].set_xlabel('x-axis')
axes[0, 0].set_ylabel('y-axis')

axes[0, 1].plot(x, y2, 'r')
axes[0, 1].set_title('Cosine Function')
axes[0, 1].set_xlabel('x-axis')
axes[0, 1].set_ylabel('y-axis')

axes[1, 0].plot(x, y3, 'g')  
axes[1, 0].set_title('Tangent Function')
axes[1, 0].set_xlabel('x-axis')
axes[1, 0].set_ylabel('y-axis')

axes[1, 1].plot(x, y4, 'm')  
axes[1, 1].set_title('Exponential Function')
axes[1, 1].set_xlabel('x-axis')
axes[1, 1].set_ylabel('y-axis')

fig.tight_layout()
plt.show()
```
